```python
x=df.iloc[:,1:2]        # In pandas, iloc stands for "index location."
y=df.iloc[:,-6:-5] 
# This selects all rows (:) and the column at integer location -6 (the sixth column from the end) up to,
# but not including, the column at integer location -5 (the fifth column from the end)
```

```python
from sklearn.model_selection import train_test_split 
#  used to split a dataset into training and testing sets
# train_test_split function from the sklearn.model_selection module in scikit-learn is a utility 
# function that splits a dataset into two or more subsets
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(x , y , test_size=0.25)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
Reg=LogisticRegression()
```

```python
Reg.fit(X_train , Y_train)
```

```python
y_pred=Reg.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score


# In[43]:


print(accuracy_score(Y_test,y_pred))
```

```python
from sklearn.metrics import classification_report


# In[45]:


from sklearn.metrics import confusion_matrix


# In[46]:


import seaborn as sns


# In[47]:

print("This is")
print(classification_report(Y_test,y_pred, zero_division=1))



print(confusion_matrix(Y_test,y_pred))
sns.heatmap(confusion_matrix(Y_test,y_pred),annot=True)
```
